```javascript
class Persona {
    constructor(nombre, apellido) {
        this.nombre = nombre;
        this.apellido = apellido;
        this._edad = 0; // propiedad adicional
    }

    get nombreCompleto() {
        return `${this.nombre} ${this.apellido}`;
    }

    get edad() {
        return this._edad;
    }

    set edad(valor) {
        if (valor < 0) {
            console.error("La edad no puede ser negativa");
        } else {
            this._edad = valor;
        }
    }

    toString() {
        return `Nombre: ${this.nombre}, Apellido: ${this.apellido}, Edad: ${this.edad}`;
    }
}

let ana = new Persona("Ana", "Martínez");
ana.edad = 25; // establecer la edad
ana.edad++;

console.log(ana.toString()); 
```

```javascript

class Producto{
    static _proximoId = 1;

    static crearId() {
        return Producto._proximoId++;
    }

    constructor(nombre, precio, cantidad = 0) {
        this.nombre = nombre;
        this.precio = precio;
        this.cantidad = cantidad;
        this.id = Producto.crearId();
    }

    vender(cantidad){
        if (cantidad <= 0) return;
        if (this.cantidad < cantidad) return

        this.cantidad -= cantidad;
    }

    comprar(cantidad) {
        if (cantidad <= 0) return;
        this.cantidad += cantidad;
    }

    get importe () {
        return this.precio * this.cantidad;
    }

    toString() {
        return `${this.nombre} $${this.precio} (x ${this.cantidad}) [${this.id}]`;
    }
}

class Catalogo{
    constructor() {
        this.productos = [];
    }

    agregar(producto) {
        this.productos.push(producto);
    }

    listar() {
        console.log("\n=== Listado de productos ===");
        for(let producto of this.productos) {
            console.log(producto.toString());
        }
        console.log(`Total: $${this.calcularValoracion()}`);
    }

    debeReponer(){
        // let resultado = []
        // for(let p of this.productos) {
        //     if (p.cantidad < 5) {
        //         resultado.push(p);
        //     }
        // }
        // return resultado;
        return this.productos.filter(p => p.cantidad < 5);
    }

    calcularValoracion(){
        // let vivo = this.filter(p => p.cantidad >= 5);
        // let importes = vivo.map(p => p.importe);
        // return importes.reduce((total, importe) => total + importe, 0);
        return this.productos
                    .filter(p => p.cantidad >= 5)
                    .map(p => p.importe)
                    .reduce((total, importe) => total + importe, 0);

    }
}

let cocaCola  = new Producto("Coca Cola", 1500, 10);
let pepsiCola = new Producto("Pepsi Cola", 1400, 5);

let catalogo = new Catalogo();
catalogo.agregar(cocaCola);
catalogo.agregar(pepsiCola);

catalogo.listar();

cocaCola.vender(7);
catalogo.listar();

pepsiCola.vender(50);
catalogo.listar();

console.log("\n=== Productos que deben reponerse ===");
for(let p of catalogo.debeReponer()) {
    console.log(p.toString());
}
```
